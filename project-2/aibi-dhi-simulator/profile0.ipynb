## Description

This profile represents an individual whose behavior is highly consistent, unfazed by stress or fatigue, but who lacks competitive drive or motivation to break personal records. To mirror this behaviour we implemented the following changes:
 1. Inside get_motivation()
  - Removed last_activity_score: The user does not care about their own records, so past activity successes should no longer positively impact their current motivation.
  - Removed valence and sufficient_sleep: The user does not care about being stressed or tired. Removing these variables means the user will remain consistent no matter the stress or lack of sleep.
   - Added base_motivation = 5: To counteract the previous changes, we assume the user has some level of intrinic basic, but low motivation which remains consistant.

2. Adjustments to get_ability()
 - Removed load (cognitive load): Due to the user stating he does not care about being stressed out.

3. Adjustments to get_trigger()
 - Removed good_arousal: Consistent no matter the stress/arousal so it was removed

# Intervention

```python
def intervention(method, runs, name, condition):
    rr_random_list = []
    not_random_list = []
    sleep_list = []
    positive_list = []
    performed = 0
    for run in range(0, runs):

        env1 = Patient(behavior_threshold=20, habituation=habituation, time_preference_update_step=time_preference_update_step)
        env1 = method(env1)
        if sum(env1.num_performed):
            performed += 1
        df = pd.DataFrame()
        df['response_ratio'] = env1.rr
        df['activity performed'] = env1.num_performed
        df['notifications'] = env1.num_notified
        df['sleep'] = env1.h_slept
        df['positive'] = env1.h_positive
        df['non_stationary'] = env1.h_nonstationary
        
        df.to_csv("{0}/patient1_{1}_{2}_run{3}.csv".format(results_directory, condition, name, run))
        rr_random_list.append(df.response_ratio)
        not_random_list.append(df.notifications)
        sleep_list.append(df.sleep)
        positive_list.append(df.positive)

    
    print("This figures include runs in which no prompt resulted in the activity being performed")
    print("{0} out of {1} resulted in activity being performed ".format(performed, runs))
    fig = plt.figure( figsize=(8, 5))
    plt.plot(np.nanmean(np.array(sleep_list), axis=0), label='Hours slept', color ='r')
    plt.plot(np.mean(positive_list, axis=0), label='Hours in positive mood')
    plt.ylabel('Hours ')
    plt.xlabel('Intevention Days')
    plt.legend()
    plt.show()

    fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True, sharey=False, figsize=(10, 8))
    ax1.plot(np.nanmean(rr_random_list, axis=0), label='response ratio', color ='r')
    ax2.plot(np.nanmean(not_random_list, axis=0), label='num of notifications')
    ax2.set_xlabel('Intervention days')
    ax2.set_ylabel('No. of notifications')
    ax1.set_ylabel('Response ratio')
    plt.show()
    return sleep_list, positive_list, rr_random_list, not_random_list
```

## Heuristic

Notify every hour except in night and when the patient is alseep

```python
def always_notify(env1, intervention_legth=eight_weeks):
    
    action = 0
    for i in  range(intervention_legth):
        
        observation, reward, terminated, truncated, info = env1.step(action)
        if observation[9] == 3 or observation[3] == 1: # do not prompt at night (index 9) or when sleeping (index 3)
            action = 0
        else:
            action = 1
            
    return env1
```

```python
sleep_listh, positive_listh, rr_h_list, not_h_list = intervention(always_notify, runs, 'h', condition)
```

## Random 
Randomly sample during the day skip the nights

```python
def random_notification(env1, intervention_legth=eight_weeks):
    
    action = 0
    for i in  range(intervention_legth):
        
        observation, reward, terminated, truncated, info = env1.step(action)
        if observation[9] ==3:
            action = 0
        else:
            action = env1.action_space.sample()
            
    return env1
```

```python
%time _ = intervention(random_notification, runs, 'random', condition)
```

```python
def get_rr_and_notification(name, runs, condition):
    
    rr, noti = [],[]
    failed = 0
    for run in range(runs):
        df = pd.read_csv("{0}/patient1_{1}_{2}_run{3}.csv".format(results_directory,condition, name, run))
        a_performed = sum(df['activity performed'].values)
        if a_performed > 0: # activity performed at least once in the course of the intervention
            rr.append(df.response_ratio)
            noti.append(df.notifications)
        else:
            failed = failed +1 
    print(" {0} out of {1} runs had no activity performed throughout the full intervention.".format(failed, runs))
    return rr, noti, failed
```

```python
rr_random_list, not_random_list, fppo = get_rr_and_notification('random', runs, condition)
```

## Supervised Learning

### Static model

```python
def supervised_after_three_weeks(env1):
    
    three_weeks = 24 * 7 * 3
    env1 = always_notify(env1, intervention_legth=three_weeks)
    clf = RandomForestClassifier(class_weight='balanced')
    clf.fit(env1.observation_list, env1.activity_performed)
    remaining_time = eight_weeks - three_weeks
    observation = env1._get_current_state()
    for i in range(remaining_time):
        # applying supervised model
        action = clf.predict(np.array([observation]))[0]
        observation, reward, terminated, truncated, info = env1.step(action)
        
    return env1
```

```python
%time _ = intervention(supervised_after_three_weeks,runs , 'static_sup3', condition)
```

### Adaptive model 

```python
def supervised_adaptive_after_three_weeks(env1):
    
    three_weeks = 24*7*3
    env1 = always_notify(env1, intervention_legth=three_weeks)
    clf = RandomForestClassifier(class_weight='balanced')
    clf.fit(env1.observation_list, env1.activity_performed)
    remaining_time = eight_weeks - three_weeks
    observation = env1._get_current_state()
    samples = len(env1.observation_list)
    for i in range(remaining_time):
        # applying supervised model
        action = clf.predict(np.array([observation]))[0]
        observation, reward, terminated, truncated, info = env1.step(action)
        if len(env1.observation_list)> samples: # retrain when new samples are provided
            clf = RandomForestClassifier(class_weight='balanced')
            clf.fit(env1.observation_list, env1.activity_performed)
            samples = len(env1.observation_list)
            
        
    return env1
```

```python
%time _ = intervention(supervised_adaptive_after_three_weeks, runs, 'adaptive_sup3', condition)
```

## Observations

Profile 0 which describes a consistent individual with baseline level of motivation but no desire to compete with others or themselves created unusual, but interesting results in a simulation, one of which is the statistic of happiness level being almost constant with upwards of 22 hours of the day of the patient being happy. Though, unusual it does, in a sense, match the idea of the patient being Intrinsicly satisfied. Regardless, the results obtained by the algorithms seem to suggest that the Adaptive Random Forrest strategy performed the best out of all the tested ones but at a high cost to the number of notifications sent at the start of the experiment. Another interesting observation is that both DQN and PPO end up very close in terms of performance on the activity-performed-to-prompt-ratio metric, however DQN decreases the total number of notifications sent in a day much more quickly than PPO suggesting that it might be a preferable alternative in case the Adaptive Random Forrest startegy ends up sending too many messages.

The results might be motivated by the high level of intrinsic motivation which leads the patient to quickly converge to a state of high happiness and strong response to sent notifications. Of course, due to the limited knowledge about the patient, it was assumed some level of intrinsic motivation is present, however, patient profile might be interpreted differently with all negative answers meaning no motivation and as such no affect from stress or being tired since the patient is unwilling to move anyway. Ultimately, this interpretation was rejected due to being too pessimistic and not allowing for almost any optimization since the patient might not respond positively to any stimulus.

In conclusion, Profile 0 patients may be encouraged to train through a more aggressive staretgy of the Adaptive Random Forrest, however, DQN algorithm might also be a good choice in case the number of notifications needs to be kept relatively low.
